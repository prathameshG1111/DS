Download the Iris flower dataset or any other dataset into a DataFrame. (e.g., 
https://archive.ics.uci.edu/ml/datasets/Iris ). Scan the dataset and give the inference as:
1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
3. Create a boxplot for each feature in the dataset.
4. Compare distributions and identify outliers.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# Column names for the dataset
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris = pd.read_csv(url, names=column_names)
iris.head()
```

```python
# 1. List down the features and their types
# Print features and their data types
print("Features and their types:")
print(iris.dtypes)
print()
```

```python
# 2. Create a histogram for each feature
# Histograms for numeric features
iris.hist(figsize=(10, 8))
plt.suptitle("Histograms of Iris Dataset Features")
plt.show()
```

```python
# 3. Create a boxplot for each feature
# Boxplot for each numeric feature
plt.figure(figsize=(10, 8))
sns.boxplot(data=iris.drop(columns='class'))
plt.title("Boxplot of Iris Dataset Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Compare distributions and identify outliers
# Outlier detection using boxplots
plt.figure(figsize=(10, 8))
sns.boxplot(data=iris.drop(columns='class'), orient='h', fliersize=5)
plt.title("Boxplot of Iris Dataset Features with Outliers")
plt.show()
```
