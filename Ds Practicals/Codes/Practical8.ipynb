1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information about 
the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we 
can find any patterns in the data.
2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger
is distributed by plotting a histogram.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
# Finding patterns of data. --> Patterns of data can be find out with the help of different types of plots
# A. Distribution Plots:
# 1. Distplot
sns.distplot(x = dataset['age'], bins = 10,kde=False)
```

```python
# 2. Joint Plot
# For Plot 1:
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind = 'scatter')
# For Plot 2:
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind = 'hex')
```

```python
# 3. Rug Plot
sns.rugplot(dataset['fare'])
```

```python
# B. Categorical Plots
# 1. The Bar Plot
sns.barplot(x='sex', y='age', data=dataset)
```

```python
# 2. The Count Plot
sns.countplot(x='sex', data=dataset)
```

```python
# 3. The Box Plot
sns.boxplot(x='sex', y='age', data=dataset)
sns.boxplot(x='sex', y='age', data=dataset, hue="survived") # along with the information about whether or not they survived
```

```python
 # 4. The Violin Plot --> the violin plot allows us to display all the␣components that actually correspond to the data point.
sns.violinplot(x='sex', y='age', data=dataset)
sns.violinplot(x='sex',y='age',data=dataset,hue='survived') # along with theinformation about whether or not they survived
```

```python
# C. Advanced Plots:
# 1. The Strip Plot
sns.stripplot(x='sex', y='age', data=dataset, jitter=False)
sns.stripplot(x='sex', y='age', data=dataset, jitter=True)
sns.stripplot(x='sex', y='age', data=dataset, jitter=True, hue='survived')
```

```python
# 2. The Swarm Plot
sns.swarmplot(x='sex', y='age', data=dataset)
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
# Checking how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.
sns.histplot(dataset['fare'], kde=False, bins=10)
```
